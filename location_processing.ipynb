### Subdivisions / Provinces

```python
path = 'un_loccodes/2019-2 SubdivisionCodes.csv'
```

```python
with open(path, errors='ignore') as f:
    subdivisions = StringIO(f.read())
```

```python
df = pd.read_csv(subdivisions, header=None)
```

```python
df[2]
```

```python
import re
```

Removed the Alias names in [...] for the provinces here

```python
provinces = []
for entry in df[2].unique().tolist():
    match = re.search(r'[^\[\(\/]+', entry)
    provinces.append(match.group(0))
```

```python
with open('location/provinces.txt', 'w') as f:
    for province in provinces:
        f.write(province.strip()+'\n')
```

### Countries

```python
with open('un_loccodes/countries_unclean.txt') as f:
    entries = f.read().strip().split('\n')
```

Cleaned the names like China (People's Republic) -> China

```python
countries = []
for country in entries:
    match = re.search(r'[^\[\(\/]+', country)
    countries.append(match.group(0))
```

```python
countries = sorted(list(set(countries)))
```

```python
countries
```

```python
with open('location/countries.txt', 'w') as f:
    for country in countries:
        f.write(country.strip()+'\n')
```
